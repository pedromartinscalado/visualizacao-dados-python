```python

# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt
import pandas as pd

# 02. Valore(s) do(s) gráfico(s)
blood_sample_men = [112,90,76,94,50,70,113,140,150,160,139,120,76,77,99,88,110,67,50,80]
blood_sample_women = [132,80,66,64,90,80,143,150,130,140,119,60,76,97,69,128,150,167,150,180]


# 03. Definição e configuração do gráfico
n , bins , patches =plt.hist([blood_sample_men,blood_sample_women], 
                             bins =  [80,100,125,150] , 
                             rwidth=0.95,
                             color = ['green','orange'],
                             label= ['men','women'])


# 04. Definição do Título e dos Rótulos do Eixos e Legenda
plt.title("Boold Sugar Chart")
plt.xlabel("Sugar Level")
plt.ylabel("Number of Patients")
#plt.legend()

# 05. Definição e Configuração de "patches"
plt.setp(patches[1][2], color='cyan')
plt.setp(patches[1][1], color='red')
df = pd.DataFrame(dict(data=n[0]))
#df['data'].plot(kind='bar', color='red')
df['data'].plot(kind='bar', color='cyan')
df['data'].plot(kind='line', marker='*', color='darkblue', ms=10)


# 06. Apresentação do gráfico
plt.show()



#print(patches[1][0].get_xy())
```

```python
# 01. Importar:
#     Bibliotecas: matplotlib, pandas
#     Módulos: pyplot
import matplotlib.pyplot as plt
import pandas as pd

# 02. Valore(s) do(s) gráfico(s)
blood_sample_men = [112,90,76,94,50,70,113,140,150,160,139,120,76,77,99,88,110,67,50,80]
blood_sample_women = [132,80,66,64,90,80,143,150,130,140,119,60,76,97,69,128,150,167,150,180]


# 03. Definição e configuração do gráfico
fig, ax = plt.subplots()
df = pd.DataFrame({'Nível de Açúcar':blood_sample_women})
df['Nível de Açúcar'].plot(kind='bar', color='#F08080', 
                          title="Nível de Açúcar no Sangue - Mulheres",xlabel="Amostras",
                          ylabel="Nível de Açúcar", alpha=0.6) 
df['Nível de Açúcar'].plot(kind='line', marker='*', color='red', ms=10) # Cor vermelha para a linha

plt.ylim(0, 185)

plt.grid(axis='y', linestyle='--')

plt.show()
```
