```python
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#                  statistics
#     Módulos: pyplot
import numpy as np
import matplotlib.pyplot as plt
import statistics
 

# 02. Valore(s) do(s) gráfico(s)
# Data in numpy array
d = np.array([189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191,134,144,137,139,
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167,140,130,150,190,171,189,142,153,159,161,169,165,133,188])


# 03. Definição e configuração do gráfico
plt.hist(d, bins=50 , color='c' , edgecolor = 'k')

# 04. Valore(s) Estatísticos
medianline = [x/2 for x in d]
print(medianline)
m  = statistics.mean(d)
sd =  statistics.stdev([189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191,134,144,137,139,
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167,140,130,150,190,171,189,142,153,159,161,169,165,133,188])

# 05. Adição dos valors estatísticos ao "plot"
plt.axvline(m, color= 'k' , linestyle="dashed") # Media - linha a preto
plt.axvline(m+sd, color= 'y' , linestyle="dashed") # Media + SD
plt.axvline(m-sd, color= 'y' , linestyle="dashed") # Media - SD 


# 06. Apresentação do gráfico
plt.show()
```

<h2 style="background-color:orange;color:white;padding-top:10px;text-align:center;font-weight:bold;font-style:italic;font-size:1cm;padding-bottom:20px;">  "Histogramas" - Estatistica II </h2>

```python
################################
# Cell 06:
################################

# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#                  statistics
#     Módulos: pyplot
import numpy as np
import matplotlib.pyplot as plt
import statistics
 

# 02. Valore(s) do(s) gráfico(s)
# Data in numpy array
d = np.array([189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191,134,144,137,139,
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167,140,130,150,190,171,189,142,153,159,161,169,165,133,188])


# 03. Definição e configuração do gráfico
plt.hist(d, bins=10 , color='c' , edgecolor = 'k')

# 04. Valore(s) Estatísticos
medianline = [x/2 for x in d]
print(medianline)
m  = statistics.mean(d)
md =  statistics.median(d)
sd =  statistics.stdev([189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191,134,144,137,139,
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167,140,130,150,190,171,189,142,153,159,161,169,165,133,188])

# 05. Adição dos valors estatísticos ao "plot"
plt.axvline(m, color= 'k' , linestyle="dashed") # Media - linha a preto
plt.axvline(m+sd, color= 'y' , linestyle="dashed") # Media + SD
plt.axvline(m-sd, color= 'y' , linestyle="dashed") # Media - SD 
plt.axvline(md, color= 'b' , linestyle="dashed") # Mediana

# 06. Apresentação do gráfico
plt.show()


```
