```python
#Histogram
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt
  
# 02. Valore(s) do(s) gráfico(s)
height = [189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191, 
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167]

# 03. Definição e configuração do gráfico  
plt.hist(height, orientation='horizontal',rwidth=.8, color='orange',
        density='True')

plt.title("Distribuição de Alturas da Turma")
plt.xlabel("Número de Ocorrências")
plt.ylabel("Alturas (cm)")



# 04. Apresentação do gráfico
plt.show()
```
