```python
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
# importing package
import matplotlib.pyplot as plt
import numpy as np

# 02. Valore(s) do(s) gráfico(s)
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

# 03. Definição do Estilo do Gráfico
plt.style.use('seaborn-v0_8-poster')

# 04. Definição e configuração do gráfico
# plot bars in stack manner
plt.barh(x, y1, color='r')
plt.barh(x, y2, left=y1, color='b')
plt.barh(x, y3, left=y1+y2, color='y')
plt.barh(x, y4, left=y1+y2+y3, color='g')

# 04. Título, Rótulos dos Eixos e Legenda
plt.title("Cadeia de Lojas de Pronto-a-Vestir 'Pega e Leva'")
plt.ylabel("Lojas")
plt.xlabel("Número de Itens Vendidos")
plt.legend(["Saias", "Calças", "Casacos", "Camisolas"])


# 05. Apresentação do gráfico
plt.show()
```
