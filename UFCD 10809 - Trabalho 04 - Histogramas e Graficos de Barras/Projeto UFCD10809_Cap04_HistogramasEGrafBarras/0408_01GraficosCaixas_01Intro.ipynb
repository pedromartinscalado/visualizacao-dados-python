```python
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt


# 02. Valore(s) do(s) gráfico(s)
x = [10,20,30,40,50,60,70,120]
x1 = [10,20,40,50,30,60,90]
x2 = [10,20,40,50,70,90,130]
y  = [x,x1,x2]

# 03. Definição e configuração do gráfico
# # Versão antiga
# plt.boxplot(y , notch =False , vert = True , widths = 0.2 , labels = ['PYTHON',"C++","JAVA"]
#             ,showmeans = True , boxprops = dict(color= "red") ,
#              capprops = dict(color = "red") , 
#              sym= "g+",  
#              whiskerprops = dict(color = "blue") )

# Nova Versão - Matplotlib 3.9
plt.boxplot(y , notch =False , vert = True , widths = 0.2 ,
            showmeans = True , boxprops = dict(color= "red") ,  
            tick_labels = ['PYTHON',"C++","JAVA"],capprops = dict(color = "red") , sym= "g+",  
whiskerprops = dict(color = "blue") )

# 04. Apresentação do gráfico
plt.show()
```
