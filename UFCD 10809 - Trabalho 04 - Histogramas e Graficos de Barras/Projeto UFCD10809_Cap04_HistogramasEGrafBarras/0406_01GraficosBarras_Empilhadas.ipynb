```python
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     M칩dulos: pyplot
# importing package
import matplotlib.pyplot as plt
import numpy as np


# 02. Valore(s) do(s) gr치fico(s)
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])


# 03. Defini칞칚o e configura칞칚o do gr치fico
# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')

# 04. T칤tulo, R칩tulos dos Eixos e Legenda
plt.title("Vendas da Cadeia de Pronto-a-Vestir 'Pega e Leva'")
plt.xlabel("Lojas")
plt.ylabel("N칰mero de Itens Vendidos")
plt.legend(["Saias", "Cal칞as", "Casacos", "Camisolas"])


# 05. Apresenta칞칚o do gr치fico
plt.show()
```
