```python
# Histograma de Frequência Absoluta
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt
  
# 02. Valore(s) do(s) gráfico(s)
values = [87, 53, 66, 61, 67, 68, 62,
          110, 104, 61, 111, 123, 117,
          119, 116, 104, 92, 111, 90,
          103, 81, 80, 101, 51, 79, 107,
          110, 129, 145, 139, 110]
  
# 03. Definição e configuração do gráfico
#     Cinco classes - Gráfico da frequência absoluta
plt.hist(values , bins =  5, edgecolor = "navy" , color=  "cyan")


# 04. Apresentação do gráfico
plt.show()
```

```python
# Histograma de Frequência Absoluta Acumulada
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt
  
# 02. Valore(s) do(s) gráfico(s)
values = [87, 53, 66, 61, 67, 68, 62,
          110, 104, 61, 111, 123, 117,
          119, 116, 104, 92, 111, 90,
          103, 81, 80, 101, 51, 79, 107,
          110, 129, 145, 139, 110]
  
# 03. Definição e configuração do gráfico
#     Cinco classes - Gráfico da frequência absoluta
plt.hist(values , bins =  5, edgecolor = "navy" , color=  "cyan", cumulative=True)


# 04. Apresentação do gráfico
plt.show()
```

```python
# Histograma de Frequência Relativa
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import matplotlib.pyplot as plt
import numpy as np
  
# 02. Valore(s) do(s) gráfico(s)
values = [87, 53, 66, 61, 67, 68, 62,
          110, 104, 61, 111, 123, 117,
          119, 116, 104, 92, 111, 90,
          103, 81, 80, 101, 51, 79, 107,
          110, 129, 145, 139, 110]
  
# 03. Definição e configuração do gráfico
#     Cinco classes - Gráfico da frequência absoluta
plt.hist(values , bins =  5, edgecolor = "navy" , color=  "cyan",weights=np.ones_like(values) / len(values))


# 04. Apresentação do gráfico
plt.show()
```

```python
# 01. Importar:
#     Bibliotecas: matplotlib, numpy
#     Módulos: pyplot
import numpy as np
import matplotlib.pyplot as plt
 
# 02. Valore(s) do(s) gráfico(s)
# Data in numpy array
exp_data = np.array([189, 185, 195, 149, 189, 147, 154, 
          174, 169, 195, 159, 192, 155, 191,134,144,137,139,
          153, 157, 140, 144, 172, 157, 181, 
          182, 166, 167,140,130,150,190,171,189,142,153,159,161,169,165,133,188])

 
# 03. Definição e configuração do gráfico
# Plot the distribution of numpy data
plt.hist(exp_data, bins = 6 , align='left', color='orange', edgecolor='brown',linewidth=1)
 

# 04. Definição dos Rótulos
plt.title("Exemplo de Histograma - Frequência Absoluta")
# Add axis labels
plt.xlabel("Alturas em cm")
plt.ylabel("Número de Estudantes")


# 05. Definição dos valores dos eixos
#     Eixos dos XX: de 0 até 19 variando de 2 em 2.
#     Eixos dos YY: de 0 até 19 variando de 1 em 1.
plt.xticks(range(120,200,10))
plt.yticks(range(0,16,1))


# 06. Apresentação do gráfico
plt.show()
```
