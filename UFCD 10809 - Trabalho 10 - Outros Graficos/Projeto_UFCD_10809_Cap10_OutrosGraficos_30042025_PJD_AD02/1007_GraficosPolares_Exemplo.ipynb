```python
import matplotlib.pyplot as plt
import numpy as np

theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(3*theta)
# r = np.sin(5*theta)

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r)
plt.title("Polar Chart Example")
plt.show()
```
