```python
import pandas as pd

"""
Célula 04: Extração dos dados do ficheiro para um dataframe e apresentação das 10 primeiras linhas.
"""

# Nome do ficheiro Excel
nome_ficheiro = "OrdensBolachas.xlsx"

# Variável para guardar o dataframe
df_bolachas = None

# --- Extrair os dados para um dataframe ---
try:
    # Usamos pd.read_excel() porque o ficheiro é .xlsx
    df_bolachas = pd.read_excel(nome_ficheiro)
    print(f"Ficheiro '{nome_ficheiro}' carregado com sucesso para o dataframe.")

except FileNotFoundError:
    print(f"Erro: O ficheiro '{nome_ficheiro}' não foi encontrado.")
    print("Certifique-se de que o ficheiro está na mesma pasta que o notebook.")
except Exception as e:
    print(f"Ocorreu um erro ao carregar o ficheiro: {e}")

# --- Apresentar as primeiras dez linhas do dataframe, se tiver sido carregado ---
if df_bolachas is not None:
    print("\nPrimeiras 10 linhas do dataframe:")
    # O método .head(n) mostra as primeiras n linhas. Por omissão, n=5.
    print(df_bolachas.head(10))
else:
    print("\nNão foi possível apresentar as primeiras linhas pois o dataframe não foi carregado devido a erro(s) acima.")
```

# 03.02 Célula 10: Aplique a função "pandas" que elimina as linhas NaN da tabela.

```python
"""
Célula 10: Eliminar linhas com valores NaN (em falta).
"""

num_linhas_antes = len(df_bolachas)
print(f"Número de linhas no dataframe ANTES de remover NaNs: {num_linhas_antes}")

# --- Aplicar a função dropna() ---
# df.dropna() remove, por omissão (axis=0), as linhas que contêm qualquer valor NaN.
# É boa prática reatribuir o resultado a uma variável, pois dropna() retorna um novo dataframe.
# Vamos reatribuir ao dataframe original 'df_bolachas'.
df_bolachas = df_bolachas.dropna().copy() # Usar .copy() é recomendado após operações de filtragem/limpeza

num_linhas_depois = len(df_bolachas)
print(f"Número de linhas no dataframe DEPOIS de remover NaNs: {num_linhas_depois}")

# Mostrar quantas linhas foram removidas
linhas_removidas = num_linhas_antes - num_linhas_depois
print(f"Total de linhas removidas: {linhas_removidas}")

print("\nPrimeiras 5 linhas do dataframe após limpeza:")
print(df_bolachas.head())
```

################################
# 04.02 Célula 13: Acrescente ao “dataframe” uma nova coluna com o nome "Year" que extrai e apresenta o ano da coluna "Date".
################################

```python
"""
Célula 13: Acrescentar coluna 'Year' extraindo o ano da coluna 'Date'.
"""

# --- Passo 1 (Importante): Garantir que a coluna 'Date' é do tipo datetime ---
# Mesmo que o pandas a tenha inferido corretamente ao carregar, é seguro garantir.
df_bolachas['Date'] = pd.to_datetime(df_bolachas['Date'])
print("Coluna 'Date' convertida (ou verificada) como tipo datetime.")

# --- Passo 2: Extrair o ano da coluna 'Date' e criar a nova coluna 'Year' ---
df_bolachas['Year'] = df_bolachas['Date'].dt.year

print("Nova coluna 'Year' criada.")

# --- Apresenta as primeiras linhas do dataframe com a nova coluna 'Year' para verificar ---
print("\nPrimeiras 5 linhas do dataframe com a nova coluna 'Year':")
print(df_bolachas.head())

# Apresenta as últimas 5 linhas também para ver se a coluna aparece no final
print("\nÚltimas 5 linhas do DataFrame:")
print(df_bolachas.tail())
```

################################
# 04.03 Célula 14: Acrescente ao “dataframe” uma nova coluna com o nome "Month" que extrai e apresenta o mês da coluna "Date".
################################

```python
"""
Célula 14: Acrescentar coluna 'Month' extraindo o mês da coluna 'Date'.
"""

# --- Extrair o mês da coluna 'Date' e criar a nova coluna 'Month' ---
# Usamos o acessor .dt para aceder a propriedades de data/hora
# .month extrai o número do mês (1 para Janeiro, 12 para Dezembro)
df_bolachas['Month'] = df_bolachas['Date'].dt.month

print("Nova coluna 'Month' (com o número do mês) criada.")

# --- Apresentar as primeiras linhas do dataframe com a nova coluna 'Month' para verificar ---
print("\nPrimeiras 5 linhas do dataframe com as colunas 'Year' e 'Month':")
# Vamos mostrar algumas colunas relevantes para confirmar que as novas colunas foram adicionadas
colunas_para_mostrar = ['Customer ID', 'Order ID', 'Date', 'Year', 'Month', 'Product', 'Units Sold', 'Revenue', 'Cost']
print(df_bolachas[colunas_para_mostrar].head())


print("\nÚltimas 5 linhas do DataFrame:")
print(df_bolachas[colunas_para_mostrar].tail())
```

```python
import matplotlib.pyplot as plt
import pandas as pd 
"""
Célula 18: Apresentar um gráfico de barras com o valor total das vendas por produto.
"""

%matplotlib inline 

# --- Cria o gráfico de barras ---
plot = df_vendas_por_produto.plot(kind='bar', x='Product', y='Total Revenue', legend=False, figsize=(10, 6))


# --- Elementos do gráfico ---
plt.title('Total de Vendas por Tipo de Produto', fontsize=14) # Título do gráfico com tamanho de fonte
plt.xlabel('Tipo de Bolacha', fontsize=12) # Rótulo do eixo X com tamanho de fonte
plt.ylabel('Total de Vendas (€)', fontsize=12) # Rótulo do eixo Y com tamanho de fonte

plt.xticks(rotation=45, ha='right')

# Ajusta automaticamente o layout da figura para que tudo caiba, incluindo rótulos
plt.tight_layout()

plt.show()
```
